# Analisis Plot

```python
plot_df = df.copy()
plot_df.columns = plot_df.columns.str.replace("_", " ").str.title()

fig = px.parallel_coordinates(
    pd.concat([plot_df.select_dtypes(float), plot_df['Track Popularity']], axis=1),
    color="Track Popularity",
    color_continuous_scale=px.colors.diverging.Spectral,
    color_continuous_midpoint=5,
    labels={'Track Popularity': 'Popularity'},
)
fig.update_layout(
    title={
        'text': "Popularitas terhadap fitur-fitur",
        'x': 0.5,
        'y': 0.99,
        'xanchor': 'center'
    },
    margin=dict(t=80)
)
fig.show()
```

## Popularitas Genre dan Subgenre

```python
plot_df[['Playlist Genre', 'Playlist Subgenre']] = plot_df[['Playlist Genre', 'Playlist Subgenre']] \
                                                  .apply(lambda x: x.str.capitalize(), axis=1)

fig = px.sunburst(
    plot_df,
    path=['Playlist Genre', 'Playlist Subgenre'],
    color='Track Popularity',
    labels={'Track Popularity': 'Popularity'},
)
fig.update_traces(
    hovertemplate="<b>%{label}</b><br><br>Count=%{value}<br>Popularity=%{color:.2f}<extra></extra>"
)
fig.update_layout(
    title={
        'text':"Popularitas Genre dan Subgenre",
        'x': 0.47,
        'y': 0.95,
        'xanchor': 'center'
    },
    margin=dict(t=70)
)
fig.show()
```

```python
# Mengkapitalisasi kolom Playlist Genre dan Playlist Subgenre
plot_df[['Playlist Genre', 'Playlist Subgenre']] = plot_df[['Playlist Genre', 'Playlist Subgenre']] \
    .apply(lambda x: x.str.capitalize(), axis=1)

# Hitung Count dan rata-rata Track Popularity untuk setiap Playlist Genre
genre_info = plot_df.groupby('Playlist Genre').agg(
    Count=('Playlist Genre', 'size'),
    Popularitas=('Track Popularity', 'mean')
).reset_index()

# Hitung Count dan rata-rata Track Popularity untuk setiap Playlist Subgenre
subgenre_info = plot_df.groupby('Playlist Subgenre').agg(
    Count=('Playlist Subgenre', 'size'),
    Popularitas=('Track Popularity', 'mean')
).reset_index()

# Tampilkan data keterangan untuk Playlist Genre
genre_info = genre_info.sort_values(by='Popularitas', ascending=False)
print("Keterangan untuk Playlist Genre:")
print(genre_info)

# Tampilkan data keterangan untuk Playlist Subgenre
subgenre_info = subgenre_info.sort_values(by='Popularitas', ascending=False)
print("\nKeterangan untuk Playlist Subgenre:")
print(subgenre_info)
```

# Lagu-lagu terpopuler berdasarkan Artis

```python
fig = px.treemap(
    plot_df.groupby(['Track Artist', 'Track Name'])['Track Popularity'].mean().sort_values(ascending=False).reset_index()[:100],
    path=[px.Constant("Artists"), 'Track Artist', 'Track Name'],
    values='Track Popularity',
    color_discrete_sequence=px.colors.diverging.Portland,
    labels={'Track Popularity': 'Popularity'},
)
fig.update_traces(
    hovertemplate="<b>%{label}</b><br><br>Popularity=%{value:.2f}<extra></extra>"
)
fig.update_layout(
    title={
        'text':"Lagu-lagu terpopuler berdasarkan Artis",
        'x': 0.5,
        'y': 0.95,
        'xanchor': 'center'
    },
    margin=dict(t=70)
)
fig.show()
```

```python
import pandas as pd

def analyze_track_popularity(df):
    top_tracks = (plot_df.groupby(['Track Artist', 'Track Name'])['Track Popularity']
                 .mean()
                 .sort_values(ascending=False)
                 .reset_index()
                 .head(100))

    # Calculate artist popularity based on sum of their tracks' popularity
    artist_popularity = (top_tracks.groupby('Track Artist')['Track Popularity']
                        .sum()
                        .sort_values(ascending=False)
                        .reset_index())

    return artist_popularity, top_tracks

# Print results
def print_results(artist_popularity, top_tracks):
    print("Keterangan untuk Nama Artis:")
    print(artist_popularity.to_string(index=True,
                                    columns=['Track Artist', 'Track Popularity'],
                                    header=['Track Artist', 'Track Popularity']))

    print("\nKeterangan untuk Judul Lagu:")
    print(top_tracks.to_string(index=True,
                             columns=['Track Artist', 'Track Name', 'Track Popularity'],
                             header=['Track Artist', 'Track Name', 'Track Popularity']))

# Example usage:
artists, tracks = analyze_track_popularity(df)
print_results(artists, tracks)
```

# Distribusi Lagu berdasarkan Tahun Rilis

```python
import pandas as pd
import plotly.express as px

# Mengubah kolom 'track_album_release_date' menjadi tahun
# Pastikan `data` sudah terdefinisi sebelumnya sebagai DataFrame
data['year'] = pd.to_datetime(data['track_album_release_date'], errors='coerce').dt.year

# Menghitung jumlah lagu per tahun
yearly_tracks = data['year'].value_counts().sort_index()

# Menghapus tahun terakhir jika datanya belum lengkap
yearly_tracks = yearly_tracks[yearly_tracks.index < 2023]  # Sesuaikan dengan dataset Anda

# Menghaluskan tren menggunakan rolling average
yearly_tracks_smoothed = yearly_tracks.rolling(window=3, center=True).mean()

# Membatasi jumlah maksimum lagu menjadi 5000
yearly_tracks_smoothed = yearly_tracks_smoothed.clip(upper=5000)

# Membuat DataFrame untuk plotly
plot_df = yearly_tracks_smoothed.reset_index()
plot_df.columns = ['Year', 'Number of Songs']

# Menambahkan kolom kategori untuk pewarnaan era
plot_df['Era'] = plot_df['Year'].apply(lambda x: 'Before 2000' if x < 2000 else '2000 and After')

# Membuat plot dengan Plotly Express
fig = px.line(
    plot_df,
    x='Year',
    y='Number of Songs',
    color='Era',
    line_group='Era',
    title='Distribusi Lagu berdasarkan Tahun Rilis',
    labels={'Number of Songs': 'Number of Songs', 'Year': 'Year'},
)

# Menambahkan anotasi untuk tahun dengan jumlah maksimum
max_year = yearly_tracks.idxmax()
max_value = yearly_tracks[max_year]
max_value = min(max_value, 5000)  # Pastikan anotasi juga mengikuti batasan maksimum
fig.add_annotation(
    x=max_year,
    y=max_value,
    text=f"Maximum in {max_year}",
    showarrow=True,
    arrowhead=2,
    ax=-50,  # Jarak horizontal dari titik anotasi
    ay=-50   # Jarak vertikal dari titik anotasi
)

fig.update_layout(
    xaxis_title='Year',
    yaxis_title='Number of Songs',
    legend_title='Era',
    margin=dict(t=50, l=25, r=25, b=25)
)

fig.show()
```

```python
# Menampilkan distribusi jumlah lagu berdasarkan tahun rilis
pd.set_option('display.max_rows', None)  # Menampilkan semua baris
print("Distribusi jumlah lagu berdasarkan tahun rilis:")
print(yearly_tracks.sort_values(ascending=False))
```

# Tren Popularitas Selama Bertahun-tahun

```python
import pandas as pd
import plotly.express as px
import numpy as np
from statsmodels.nonparametric.smoothers_lowess import lowess

# Mengubah kolom 'track_album_release_date' menjadi tahun
# Pastikan `data` sudah terdefinisi sebelumnya sebagai DataFrame
data['year'] = pd.to_datetime(data['track_album_release_date'], errors='coerce').dt.year

# Menghitung rata-rata popularitas lagu per tahun
yearly_popularity = data.groupby('year')['track_popularity'].mean()

# Menghaluskan tren menggunakan LOWESS
smoothed = lowess(yearly_popularity.values, yearly_popularity.index, frac=0.1)

# Membuat DataFrame untuk plotly
popularity_df = pd.DataFrame({
    'Year': yearly_popularity.index,
    'Average Popularity': yearly_popularity.values
})

# Membuat plot dengan Plotly Express
fig = px.line(
    popularity_df,
    x='Year',
    y='Average Popularity',
    title='Tren Popularitas Selama Bertahun-tahun',
    labels={'Average Popularity': 'Average Popularity', 'Year': 'Year'},
)

# Menambahkan tren yang dihaluskan
fig.add_scatter(
    x=smoothed[:, 0],
    y=smoothed[:, 1],
    mode='lines',
    name='Smoothed Trend',
    line=dict(color='red')
)

# Menambahkan anotasi untuk puncak dan lembah
max_year = yearly_popularity.idxmax()
max_value = yearly_popularity.max()
fig.add_annotation(
    x=max_year,
    y=max_value,
    text=f"Highest in {max_year}",
    showarrow=True,
    arrowhead=2,
    ax=-50,
    ay=50
)

min_year = yearly_popularity.idxmin()
min_value = yearly_popularity.min()
fig.add_annotation(
    x=min_year,
    y=min_value,
    text=f"Lowest in {min_year}",
    showarrow=True,
    arrowhead=2,
    ax=-50,
    ay=-50
)

fig.update_layout(
    xaxis_title='Year',
    yaxis_title='Average Popularity',
    margin=dict(t=50, l=25, r=25, b=25)
)

fig.show()
```

```python
# Menampilkan tren popularitas selama bertahun-tahun
pd.set_option('display.max_rows', None)  # Menampilkan semua baris tanpa terpotong
print("Distribusi jumlah lagu berdasarkan tahun rilis:")
print(yearly_popularity.sort_values(ascending=False))
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Baca dataset
df = pd.read_csv("spotify_songs.csv")

# Hitung rata-rata fitur per genre
features = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
           'instrumentalness', 'liveness', 'valence', 'tempo']

genre_features = df.groupby('playlist_genre')[features + ['track_popularity']].mean()
correlations = genre_features.corr()['track_popularity'].sort_values(ascending=False)

# Plot correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(genre_features.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlations by Genre')
plt.show()
```
